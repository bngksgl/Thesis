```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
def get_dataset(path, k='/Results/'):
    # qbc
    df=pd.read_csv(os.getcwd()+k+path)

    total1=[]
    for i in df['label']:
        a=i.replace('[','').replace(']','').split(',')
        total1.append(sum([int(k.strip()) for k in a]))
    df['total1s']=total1

    sm=0
    ls=[]
    for i in df['total1s']:
        sm=sm+i
        ls.append(sm)

    df['total1s_cum']=ls

    return df
```

```python
df=get_dataset('/Faster_HeadPhone_opt3_onlylev_uncertainity_svm_10', k='')
df.head()
```

```python
df.head(40)
```

```python
max(df['total1s_cum'])
```

```python
df['recall']=df['tp']/(df['tp']+df['fn'])# how many real positives i have catched
df['precision']=df['tp']/(df['tp']+df['fp'])# among the predicted matches, how many of them are really positive
df['ratio_of1s']=df['total1s_cum']/df['quota']
```

```python
df.head(40)
```

```python
plt.plot(df['quota'], df['f1_score'], 'forestgreen',label='f1 score')
plt.plot(df['quota'], df['recall'], 'lightseagreen',label='recall')
#plt.plot(df['quota'], [0.53] * (len(jac_4['quota'])), '-r',linewidth=2, label='f1 score-all data')
plt.plot(df['quota'], df['precision'], 'gold',linestyle = ':',label='precision')

plt.title('Missing option 4')
plt.xlabel('Labeled query instances')
plt.ylabel('F1 Score on test')
plt.legend(loc='lower right',prop={'size': 7.5})
plt.ylim(0, 1.2)
plt.xlim(0, 1000)
```
